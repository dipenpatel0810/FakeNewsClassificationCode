```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import os

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
import nltk
from sklearn.feature_extraction.text import TfidfTransformer, CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
import string

# Any results you write to the current directory are saved as output.
```

```python
from time import time
```

```python
train=pd.read_csv('../input/train.csv')
test=pd.read_csv('../input/test.csv')

train = train.fillna('')
test = test.fillna('')
```

```python
list(train.columns.values)
```

```python
data_train = pd.DataFrame()
data_train['all_data'] = train['title']+" " + train['author']+ " " + train['text']
data_train_labels = train['label']
```

```python
data_train
```

```python
punctuation_string = string.punctuation
stop_words = set(stopwords.words('english'))
```

```python
t1 = time()
tokenized_news_list = []
for each_news in data_train['all_data'].values:
    tokenized_news = word_tokenize(str(each_news))
    tokenized_news_list.append(tokenized_news)
```

```python
stop_words_removed = list()
for each_list in tokenized_news_list:
    removed_news = list()
    for each_word in each_list:
        each_word = each_word.lower()
        if each_word not in stop_words and each_word not in punctuation_string and each_word not in 'â€™':
            removed_news.append(each_word)
    stop_words_removed.append(removed_news)
```

```python
transformer = TfidfTransformer(smooth_idf=False)
count_vectorizer = CountVectorizer(ngram_range=(1, 2))
counts = count_vectorizer.fit_transform(data_train['all_data'].values)
tfidf = transformer.fit_transform(counts)
```

```python
X_train, X_test, y_train, y_test = train_test_split(tfidf, data_train_labels, random_state=0, test_size = 0.30)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
nbrs = KNeighborsClassifier()
nbrs.fit(X_train, y_train)
```

```python
print('Accuracy of KNN classifier on test set: {:.2f}'
     .format(nbrs.score(X_test, y_test)))
```

```python
prediction = nbrs.predict(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB

NB = MultinomialNB()
NB.fit(X_train, y_train)
print('Accuracy of NB  classifier on training set: {:.2f}'
     .format(NB.score(X_test, y_test)))
```

```python
NB = MultinomialNB(alpha = 0.1)
NB.fit(X_train, y_train)
print('Accuracy of NB  classifier on training set: {:.2f}'
     .format(NB.score(X_test, y_test)))
```

```python
NB = MultinomialNB(alpha = 0.01)
NB.fit(X_train, y_train)
print('Accuracy of NB  classifier on training set: {:.2f}'
     .format(NB.score(X_test, y_test)))
```

```python
NB = MultinomialNB(alpha = 0.001)
NB.fit(X_train, y_train)
print('Accuracy of NB  classifier on training set: {:.2f}'
     .format(NB.score(X_test, y_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
accuracy = []
  
rf= RandomForestClassifier(n_estimators=7)
rf.fit(X_train, y_train)
score = rf.score(X_test, y_test)*100
accuracy.append(score)
print('Accuracy of randomforest classifier on test set: {:.2f}'
     .format(score))
```

```python
from sklearn.linear_model import LogisticRegression
```
